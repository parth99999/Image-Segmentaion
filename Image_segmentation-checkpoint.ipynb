```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("nature.png")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
original_shape=img.shape
```

```python
#Flatten the shape of the image:
all_pixels=img.reshape((-1,3))
```

```python
all_pixels.shape
```

```python
from sklearn.cluster import KMeans
```

```python
dominant_colors=4
km=KMeans(n_clusters=dominant_colors)
km.fit(all_pixels)
```

```python
centers=km.cluster_centers_
```

```python
centers
```

```python
centers=np.array(centers,dtype="uint")
```

```python
centers
```

```python
i=1
plt.figure(0,figsize=(8,2))
colors=[]
for each_col in centers:
    plt.subplot(1,10,i)
    plt.axis("off")
    i+=1
    
    colors.append(each_col)
    
    a=np.zeros((100,100,3),dtype="uint")
    a[:,:,:]=each_col
    plt.imshow(a)
```

```python
new_img=np.zeros((413*550, 3),dtype="uint")
```

```python
new_img.shape
```

```python
for ix in range(new_img.shape[0]):
    new_img[ix]=colors[km.labels_[ix]]
new_img=new_img.reshape((original_shape))
plt.imshow(new_img)
plt.show()
```
